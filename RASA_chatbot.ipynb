```python
import rasa
```

```python
config = "config.yml"
training_files = "data/"
domain = "domain.yml"
output = "models/"
```

```python
model_path = rasa.train(domain, config, [training_files], output)
```

```python
!rasa shell
```

```python
type(model_path.model)
```

```python
import asyncio
from rasa.core.agent import Agent
from rasa.shared.utils.io import json_to_string

agent = Agent.load(model_path.model)
```

```python
help(Agent)
```

```python
# get intent
message = "vô lý quá"

message = message.strip()
result = await agent.parse_message(
                message_data=message)
print(result)
```

```python
# get text
message = "xin chào"

message = message.strip()
result = await agent.handle_text(
                message)
print(result[0]["text"])
```

```python
import gradio as gr

async def greet(message):

    message = message.strip()
    result = await agent.handle_text(
                    message)

    return result[0]["text"]

demo = gr.Interface(fn=greet, inputs="text", outputs="text")

demo.launch(share = True)
```
